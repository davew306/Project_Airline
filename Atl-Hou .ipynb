```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
csv_2019_path = "data2019.csv"
csv_2020_path = "data2020.csv"
```

```python
#Data displayed (Filtered) according to carrire name "Delta",passengers,origin "Atl" and destination "Hou"

Delta_Atl = data_2019_df[(data_2019_df['CARRIER_NAME']=='Delta Air Lines Inc.')&(data_2019_df['ORIGIN']=='ATL')& (data_2019_df['DEST']=='HOU')& (data_2019_df['PASSENGERS']>= 1000)]
Delta_Atl.head()
```
